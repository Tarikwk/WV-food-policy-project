```python
#have to name my own columns because the header data wouldnt concat when i tried to put the datatables together
COLUMNS = [
    "Fiscal Year",
    "Avg Participation (Thousands)",
    "Avg Benefit Per Person (Dollars)",
    "Total Benefits (Millions of Dollars)",
    "All Other Costs (Millions of Dollars)",
    "Total Costs (Millions of Dollars)"
]
```

```python
rows = []
```

```python
#this loop goes through each page of the .pdf to pull the data table out and filters for only numeric values
#first time it worked but didnt pull 1982 because it was footnoted:
#Puerto Rico initiated Food Stamp operations during FY 1975 and participated through June of FY 1982. A 
#separate Nutrition Assistance Grant began in July 1982
#so have to use regex to clean it up 
with pdfplumber.open("C:/Users/user/Desktop/Cohort10/Capstone DATA/SNAP Participation and costs (national level '69-'24).pdf") as pdf:
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            for row in table:
                if len(row) == 6:
                    year_match = re.match(r"(\d{4})", row[0])
                    if year_match:
                        row[0] = year_match.group(1)   
                        rows.append(row)
```

```python
#made my dataframe
snap_nat_budget_df = pd.DataFrame(rows, columns=COLUMNS)
```

```python
#cleaned up and converted to numeric
for col in snap_nat_budget_df.columns[1:]:
    snap_nat_budget_df[col] = pd.to_numeric(snap_nat_budget_df[col].str.replace(",", ""), errors="coerce")
```

```python
snap_nat_budget_df.head()
```

```python
snap_nat_budget_df["Avg Participation (Thousands)"] = pd.to_numeric(snap_nat_budget_df["Avg Participation (Thousands)"], errors="coerce")
snap_nat_budget_df["Avg Benefit Per Person (Dollars)"] = pd.to_numeric(snap_nat_budget_df["Avg Benefit Per Person (Dollars)"], errors="coerce")
snap_nat_budget_df["Total Benefits (Millions of Dollars)"] = pd.to_numeric(snap_nat_budget_df["Total Benefits (Millions of Dollars)"], errors="coerce")
snap_nat_budget_df["All Other Costs (Millions of Dollars)"] = pd.to_numeric(snap_nat_budget_df["All Other Costs (Millions of Dollars)"], errors="coerce")
snap_nat_budget_df["Total Costs (Millions of Dollars)"] = pd.to_numeric(snap_nat_budget_df["Total Costs (Millions of Dollars)"], errors="coerce")
```
